In this notebook I am scraping novels from the Gutenberg project (https://www.gutenberg.org/). This notebook is using *Python2*!

```
!pip install gutenberg
import unicodedata
from gutenberg.acquire import load_etext
from gutenberg.cleanup import strip_headers
import numpy as np
```

Using gutenberg package, I download a single item:

```
ind = 11#1238#109;
text = strip_headers(load_etext(ind, mirror = "http://mirrors.xmission.com/gutenberg/")).strip()
print(text[:71])
print('...' + text[1000:1200] + '...')    
```

I choose indices that correspond to novels in the Gutenberg database, clean the text, cut out the beggining of the books (usually tables of content and prefaces), crop the data and write the data to the text file:

```
inds=np.arange(11,25,2).tolist() + [41,43,45,47,55,57,77,81,85]
Outfile = open('data_novels.txt', 'a+')
Outfile.truncate(0)
for ix in inds:
  try: text = strip_headers(load_etext(ix, mirror = "http://mirrors.xmission.com/gutenberg/")).strip()
  except: print('no file for index ' + str(ix)); pass
  str_text = unicodedata.normalize('NFD', text).encode('ascii', 'ignore')
  w = str_text.replace("\n", " ")
  Outfile.write(w[1000:20000] + ' ')
 
```
