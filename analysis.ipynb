Plot word maps:

```python
fig,ax=plt.subplots(1,3,figsize=[30,10])
with open('gen_novels.txt') as f:
  wordcloud = WordCloud(max_font_size=45, max_words=120, background_color="white").generate(f.read())
  ax[0].imshow(wordcloud, interpolation='bilinear')
  ax[0].set_axis_off()  
  ax[0].set_title('novels-tuned',fontsize=25)
with open('gen_news.txt') as f:
  wordcloud = WordCloud(max_font_size=45, max_words=120, background_color="white").generate(f.read())
  ax[1].imshow(wordcloud, interpolation='bilinear')
  ax[1].set_axis_off()
  ax[1].set_title('news',fontsize=25)  
with open('gen_abstracts.txt') as f:
  wordcloud = WordCloud(max_font_size=45, max_words=120, background_color="white").generate(f.read())
  ax[2].imshow(wordcloud, interpolation='bilinear')
  ax[2].set_axis_off()  
  ax[2].set_title('abstracts',fontsize=25)

plt.show()
plt.savefig('map.png')
```
